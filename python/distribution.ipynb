```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("/root/ant/atec_anti_fraud_train.csv")
test = pd.read_csv("/root/ant/atec_anti_fraud_test_b.csv")
train1 = pd.read_csv("/root/ant/atec_anti_fraud_train.csv")
train1.loc[train1['label'] == -1, 'label'] = 1
```

```python
# 训练集、测试集每个特征的可能取值的个数
for i in range(1, 297+1):
    fu_train = pd.unique(train['f' + str(i)])
    fu_test = pd.unique(test['f' + str(i)])
    print(i, len(list(fu_train)), len(list(fu_test)))
```

```python
# 训练集、测试集缺失值分布
from __future__ import division
print(len(train), len(test))
for i in range(1, 297+1):
    train_nan = train['f' + str(i)].isnull().sum()
    test_nan = test['f' + str(i)].isnull().sum()
    print('f' + str(i), train_nan, test_nan, train_nan/len(train), test_nan/len(test))
```

```python
import matplotlib.pyplot as plt
train1 = train[train.label>=0]
fe = "f21"
tmp = train1
fea = tmp.groupby(fe)['label'].size()
```

```python
train_1 = train1[train1['label'] == 1]
train_0 = train1[train1['label'] == 0]
```

```python
for i in range(1, 297+1):
    train_1_max = train_1['f' + str(i)].dropna().max()
    train_1_min = train_1['f' + str(i)].dropna().min()
    train_1_nan = train_1['f' + str(i)].isnull().sum()
    train_0_max = train_0['f' + str(i)].dropna().max()
    train_0_min = train_0['f' + str(i)].dropna().min()
    train_0_nan = train_0['f' + str(i)].isnull().sum()
    print('f' + str(i), "max:", train_1_max, train_0_max, 
          "min:", train_1_min, train_0_min, "nan:", train_1_nan, train_0_nan)
```

```python
for i in range(1, 297+1):
    train_1_mean = train_1['f' + str(i)].dropna().mean()
    train_1_median = train_1['f' + str(i)].dropna().median()
    train_1_mode = train_1['f' + str(i)].mode().values[0]
    train_0_mean = train_0['f' + str(i)].dropna().mean()
    train_0_median = train_0['f' + str(i)].dropna().median()
    train_0_mode = train_0['f' + str(i)].mode().values[0]
    print('f' + str(i), "mean:", train_1_mean, train_0_mean, 
          "median:", train_1_median, train_0_median, "mode:", train_1_mode, train_0_mode)
```
