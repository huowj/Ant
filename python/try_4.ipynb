```python
import pandas as pd
```

```python
train = pd.read_csv("/root/ant/atec_anti_fraud_train.csv")
test_b = pd.read_csv("/root/ant/atec_anti_fraud_test_b.csv")
```

```python
from sklearn.utils import shuffle
import xgboost as xgb
train.loc[train['label'] == -1, 'label'] = 1
train = shuffle(train)
```

```python
train = train.fillna(-1)
```

```python
validation = train[train["date"] >= 20171025]
train = train[train["date"] <= 20171024]
```

```python
len(validation)
```

```python
validation.drop(["id", "date"], axis = 1, inplace = True)
```

```python
validation_y = validation.label
validation_x = validation.drop(['label'],axis=1)
```

```python
params = {
    'colsample_bytree': 0.2,
    'learning_rate': 0.01,
    'objective': 'binary:logistic',
    'max_depth': 5,
    'eval_metric': 'auc',
    'silent': 0
}
```

```python
xgb_validation = xgb.DMatrix(validation_x,label=validation_y)
```

```python
train.drop(["id"], axis = 1, inplace = True)
```

```python
model1 = xgb.train(params, xgb_train1, 5000, watchlist, early_stopping_rounds=100)
```
